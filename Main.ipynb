```python
import pandas as pd
df=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
df=df.drop(["customerID"],axis=1)
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0).astype(int)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="mean")
```

```python
df.info()
```

```python
num_cols=["SeniorCitizen","tenure","MonthlyCharges","TotalCharges"]
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), # Instance
    ('scaler', StandardScaler())                   # Instance
])
```

```python
cat_cols=["gender","Partner","Dependents","PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport",
          "StreamingTV","StreamingMovies","Contract","PaperlessBilling","PaymentMethod"]
from sklearn.preprocessing import OneHotEncoder
cat_imputer = SimpleImputer(strategy="most_frequent") 

cat_pipeline = Pipeline(steps=[
    ('imputer', cat_imputer), 
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(["Churn"],axis=1)
y=df["Churn"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_cols),
        ('cat', cat_pipeline, cat_cols)
    ],
    remainder='passthrough'
)
```

```python
X_train_final=preprocessor.fit_transform(X_train)
X_test_final=preprocessor.transform(X_test)
y_train_final=y_train.map({"Yes":1,"No":0})
y_test_final=y_test.map({"Yes":1,"No":0})
```

```python
#Model training
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(X_train_final,y_train_final)
```

```python
from sklearn.metrics import f1_score
y_pred=model.predict(X_test_final)
score = f1_score(y_test_final, y_pred)
print(score)
```

```python
y_train_final.value_counts()
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mae = mean_absolute_error(y_test_final, y_pred)
print(MAE)
```

```python
import joblib
full_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', model)
])
joblib.dump(full_pipeline, 'churn_predl.pkl')
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Assuming you have trained a model (e.g., model = LogisticRegression().fit(X_train_final, y_train))
y_pred = model.predict(X_test_final)

# Calculate real accuracy
accuracy = accuracy_score(y_test_final, y_pred)
print(f"Actual Accuracy: {accuracy * 100:.2f}%")

# See precision and recall (very important for Churn!)
print(classification_report(y_test_final, y_pred))
```
